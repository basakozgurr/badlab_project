#Veri Analizi ve Veri Görselleştirme

```python
#ilgili kütüphaneleri ekliyoruz
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("imdb_top_1000.csv")
df.head() #veriyi görmek için yaptık
#df.info() #veri hakkındaki bilgileri aldık
```

```python
#kolaylık sağlaması için sütunların ismini değiştiriyoruz
df = df.rename(columns={
    'Series_Title': 'Title',
    'Released_Year': 'Year',
})
```

```python
#null değerleri kontrol edelim
df.isnull().sum()
```

```python
df.describe() #veri hakkında genel bilgileri aldık
```

```python
df['Gross'] = df['Gross'].str.replace(',', '').astype(float) #kolaylık sağlaması için tür değiştiriyoruz

df['Runtime'] = df['Runtime'].str.replace('min', '').astype(int)

df.info() #türlerini değiştirebildik mi diye kontrol etme
```

```python
df.dropna(subset= ['Gross'], inplace=True) #eksik değerleri siler
```

```python
#eksik veriler model performansını düşürebileceği için eksik verilerin olduğu yeri
#dolduracağız ve dolduracağımız kısmı mod alıp eç çok tekrar eden veri ile yapacağız
df['Certificate'] = df['Certificate'].fillna(df["Certificate"].mode()[0])
df['Meta_score'] = df['Meta_score'].fillna(df["Meta_score"].mode()[0])
df.shape
```

```python
#gereksiz verileri siliyoruz poster linki gibi
df = df.drop(['Poster_Link'], axis=1)
df
```

```python
#Hasılatı en yüksek olan 10 filmi bulalım
top_10_gross = df.nlargest(10, 'Gross')
top_10_gross
```

```python

top_10_imbd = df.nlargest(10, 'IMDB_Rating')
top_10_imbd
```

```python
#bar plot ile en yüksek 10 yönetmeni sıralıyoruz:
top_directors = df['Director'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title('Top 10 Yönetmen')
plt.show()
```

```python
#saçılım grafiği yapıyoruz imdb puanı ve hasılat arasındaki ilişki için
plt.figure(figsize=(10, 6))
sns.scatterplot(x='IMDB_Rating', y='Gross', data=df)
plt.title('Hasılat vs. IMDB Puanı')
plt.show()
```

```python
#en yüksek bütçeli 10 film için çizgi grafiği
fig, axes = plt.subplots()
grouped = df.groupby("Series_Title")
mean = pd.DataFrame(grouped["Gross"].mean())
mean1 = mean.sort_values("Gross", ascending=False)
sns.barplot(x=mean1.index[:10], y=mean1.iloc[:10, 0].values, ax=axes, palette='Set3')
for container in axes.containers:
    axes.bar_label(container, label_type="center", rotation=90)
    axes.set_xticklabels(axes.get_xticklabels(), rotation=90)
plt.ylabel('Gross')
plt.show()
```
